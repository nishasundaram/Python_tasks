Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
# code here
# in the next notebook, we will learn how we can clean the "Trans" column

manual=cars.loc[cars['Transmission'] == cars.Transmission.str.contains('Man.')]
automatic=cars.loc[cars['Transmission'] == cars.Transmission.str.contains('Auto.')]
print(manual, automatic)
```

Read the `car_brands.csv` data:

```python
# code here

car_brands = pd.read_csv('data/car_brands.csv')
```

Join the cars dataframe with the car brands dataframe.

```python
# code here

joined = cars.join(car_brands.set_index('brand'), on='Make')
joined.head(10)

```

Which brands have the most revenue?

```python
# code here
#revenue = (cars.Make.value_counts())

#revenue = car_brands.groupby('brand')['revenue'].sum()
#a=revenue.sort_values(ascending = False)
#print(a)

revenue = joined.groupby('Make')['revenue'].max()
a=revenue.sort_values(ascending = False)
print(a)
```
